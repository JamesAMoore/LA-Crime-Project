## Question 1: Which Los Angeles areas have the highest amount of reported crime?

## Question 2: What are the most frequently reported crimes?

```python
df['Crm Cd Desc'].value_counts().head(5)
```

```python
# Determine the top 10 crimes areas
top_crime_areas = df['AREA NAME'].value_counts().head(10).index

# Filter the data to include only the top 10 crime areas
data_top_crime = df[df['AREA NAME'].isin(top_crime_areas)]

# Determine the top 5 most common crimes
top_crimes = data_top_crime['Crm Cd Desc'].value_counts().head(5).index

# Filter the data to include only the top 5 crimes
data_top_crimes = data_top_crime[data_top_crime['Crm Cd Desc'].isin(top_crimes)]

# Determing the order of areas based on the most common crimes
area_order = data_top_crime.groupby('AREA NAME')['Crm Cd Desc'].count().sort_values(ascending = False).head(10).index

# Create the count plot
plt.figure(figsize=(10,6))
ax = sns.countplot(data=data_top_crimes, x='AREA NAME', hue='Crm Cd Desc', palette='plasma', order=area_order)

# Customize legend to decrease size
legend = ax.legend(title='Crimes')
custom_legend_labels = legend.texts
custom_legend_labels[2].set_text('Assault w/ Deadly Weapon')
custom_legend_labels[4].set_text('Vandalism - Felony $400+')

# Adjust y-axis values 
ax.set_ylim(1000, 7500)


plt.xlabel('Location')
plt.ylabel('Cases Reported')
plt.title('Crime Distribution for Top 10 Areas')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

__Stolen Vehicle__ was the most reported crime in every area except for __Central and Hollywood__. __Burglary from Vehicle__ was the most reported in __Central__ and __Battery - Simple Assault__ was the most reported crime in __Hollywood__.

## Question 4: What are the trends in the reported crimes between 2020-2023

1. What are the trends in the categories of reported crimes?
2. Did crime rates increase or decrease?

### 1. What are the categorical trends of reported crimes?

```python
# Determine the top 5 most common crimes overall
top_crimes = df['Crm Cd Desc'].value_counts().head(5).index

# Filter the data to include only the top 5 crimes
data_top_crimes = df[df['Crm Cd Desc'].isin(top_crimes)]

data_top_crimes = data_top_crimes.sort_values(by=['year', 'Crm Cd Desc'], ascending=[True,False])

# Create the countplot
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=data_top_crimes, x='year', hue='Crm Cd Desc', palette='plasma')

legend = ax.legend(title='Crimes')
custom_legend_labels = legend.texts
custom_legend_labels[1].set_text('Vandalism - Felony $400+')

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Distribution of Top 5 Crimes per Year')

plt.tight_layout()
plt.show()
```

__Stolen Vehicle__ has been the most reported crime every year from 2020 until now. With the second most reported being __Battery - Simple Assault__.

### 2. Did crime rates increase or decrease?

```python
# Filter out 2023 because year isn't completed, could misrepresent the trend
filtered_df = df[df['year'] != 2023]

# Extract the unique years and their counts
yearly_counts = filtered_df['year'].value_counts().reset_index()
yearly_counts.columns = ['Year', 'Count']

# Sort the DataFrame by 'Year' to ensure the correct order on the x-axis
yearly_counts = yearly_counts.sort_values(by='Year')

plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_counts, x='Year', y='Count')

plt.xticks(yearly_counts['Year'].astype(int))
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Crime Trend from 2020-2023')

plt.tight_layout()
plt.show()
```

We are able to see that crime has __increased__ year over year in Los Angeles since 2020.

```python
df['Month'] = df['DATE OCC'].dt.month
df['Day of Week'] = df['DATE OCC'].dt.dayofweek

# Determine the top 5 most common crimes overall
top_crimes = df['Crm Cd Desc'].value_counts().head(5).index

# Filter the data to include only the top 5 crimes
data_top_crimes = df[df['Crm Cd Desc'].isin(top_crimes)]

data_top_crimes = data_top_crimes.sort_values(by=['year', 'Crm Cd Desc'], ascending=[True, False])

# Map the 'Day of Week' column to day names
dmap = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thurs', 4: 'Fri', 5: 'Sat', 6: 'Sun'}
data_top_crimes['Day of Week'] = data_top_crimes['Day of Week'].map(dmap)

day_order = ['Mon', 'Tue', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']

# Create the countplot
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Day of Week', data=data_top_crimes, hue='Crm Cd Desc', palette='plasma', order=day_order)

legend = ax.legend(title='Crimes', bbox_to_anchor=(1.05, 1), loc='upper left')
custom_legend_labels = legend.texts
custom_legend_labels[1].set_text('Vandalism - Felony $400+')

plt.xlabel('Day of Week')
plt.ylabel('Crimes Reported')
plt.title('Top 5 Crimes by Day of Week')

plt.tight_layout()
plt.show()
```

We can see that __Friday__ has the highest reports of __Stolen Vehicles__ just in time to inconvenience your weekend. 

__Saturday__ and __Sunday__ have the highest reports of __Simple Assaults__ taking place.
